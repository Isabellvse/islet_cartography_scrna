# Description of the AnnData object

#### Load libraries

````python
import anndata as ad
````

````python
adata = ad.read_h5ad("/work/hackaton_icrna/data/anndata/AH_combined.h5ad")
````

**AnnData object overview**

**Shape**

```python
adata.shape
```

* Cells (`n_obs`): 502,863
* Genes (`n_vars`): 60,656

**Cell metadata**

```python
adata.obs
```

Contains per-cell annotations such as sample, donor, disease status, sequencing metadata, and cell annotations.

**Gene metadata**

```python
adata.var
```

Contains gene annotations (e.g. `gene_symbol`, `ensembl_id`).

**Expression matrices**

Raw counts (Counts from full-length transcript protocols have been normalized by gene length.):

```python
adata.layers["counts"]
```
Normalized expression:

```python
adata.X
```

Log-normalized counts generated with:

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

**Embeddings**

```python
adata.obsm
```

* `X_latent_1`
* `X_umap`

**Graph structure**

```python
adata.obsp
```

* `connectivities`
* `distances`

**Analysis metadata**

```python
adata.uns
```

Contains analysis outputs such as neighbors, UMAP settings, Leiden clustering, and plotting colors.
