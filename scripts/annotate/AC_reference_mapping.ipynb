```python
# Path and system utilities
import os                    # Operating system interface
import sys                   # System-specific parameters and functions
import glob                  # File pattern matching
from pathlib import Path     # Object-oriented filesystem paths
from pyhere import here      # Reproducible project paths

# Single-cell data handling
import anndata as ad            # Core data structure for single-cell data
import scanpy as sc

# Reference mapping
import panhumanpy as ph
```

Parameters

```python
# Paths
base_dir = str(here('data/annotate/'))
plot_dir = os.path.join(base_dir, 'plot') 
files_dir = os.path.join(base_dir, 'files') 

anndata_dir = str(here('data/anndata/'))
```

Load data

```python
adata = ad.read_h5ad(os.path.join(anndata_dir, "AG_combined.h5ad"))
```

```python
azimuth = ph.AzimuthNN_base()
azimuth.query_adata(adata)
azimuth.process_query()
azimuth.run_inference_model()
_ = azimuth.process_outputs()
# refinement of labels for consistent granularity and ease of interpretation
_ = azimuth.refine_labels(refine_level = 'broad')
_ = azimuth.refine_labels(refine_level = 'medium')
_ = azimuth.refine_labels(refine_level = 'fine')
azimuth.update_cells_meta()
cell_metadata = azimuth.cells_meta
annotated_query = azimuth.pack_adata(os.path.join(files_dir, 'azimuth_reference.h5ad'))
```

Save annotations as csv

```python
annotation_df = annotated_query.obs[['full_hierarchical_labels', 'final_level_labels','final_level_confidence', 'full_consistent_hierarchy', 'azimuth_broad', 'azimuth_medium', 'azimuth_fine']].copy()
annotation_df.to_csv(os.path.join(files_dir, "azimuth_annotation.csv"), index_label = 'barcode')
```
